```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos ficticios para ilustrar la relación inversa entre temperatura y cantidad de piratas
temperatura = np.array([20, 23, 24, 25, 26]).reshape(-1, 1)  # Ejemplo de temperatura (en grados Celsius)
cantidad_piratas = np.array([50, 40, 30, 20, 10])  # Ejemplo de cantidad de piratas

# Crear un modelo de regresión lineal
modelo = LinearRegression()

print (temperatura.data.shape, type(temperatura))
print (cantidad_piratas.data.shape, type(cantidad_piratas))
# Entrenar el modelo
modelo.fit(temperatura, cantidad_piratas)

print ("score", modelo.score(temperatura,cantidad_piratas))

# Predecir la cantidad de piratas para las temperaturas dadas
prediccion_piratas = modelo.predict(temperatura)

# Graficar el scatter plot y la línea de regresión
plt.figure(figsize=(6, 4))
plt.scatter(cantidad_piratas,temperatura, color='blue', label='Datos')

# El ploteo del predictor se ve super raro
plt.plot(prediccion_piratas,temperatura, color='red', label='Regresión lineal')
```

```python


# El ploteo del predictor perfeto
# Crear un modelo de regresión lineal
cantidad_piratas2 = temperatura * 2
modelo = LinearRegression()
# Entrenar el modelo
modelo.fit(temperatura, cantidad_piratas2)
# Predecir la cantidad de piratas para las temperaturas dadas
prediccion_piratas2 = modelo.predict(temperatura)

plt.scatter(cantidad_piratas2,temperatura, color='black', label='Datos truc')
plt.plot(prediccion_piratas2,temperatura, color='green', label='Regresión lineal truc')

```

```python
# Calcular mínimo y máximo dinámicamente
min_temperatura = np.min(temperatura)
max_temperatura = np.max(temperatura)

# Extender el rango de temperatura dinámicamente
temperatura_extendida = np.arange(min_temperatura - 5, max_temperatura + 5, 0.1).reshape(-1, 1)

prediccion_extendida = modelo.predict(temperatura_extendida)  # Predicciones correspondientes

print (temperatura_extendida.data.shape, type(temperatura_extendida))

plt.plot(temperatura_extendida, prediccion_extendida, color='pink', label='Línea de predicción extendida')
```
