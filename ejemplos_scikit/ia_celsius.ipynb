```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de entrenamiento (Celsius y su equivalente en Fahrenheit)
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float).reshape(-1, 1)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)

modelo = LinearRegression()

modelo.fit(celsius, fahrenheit)

prediccion = modelo.predict(celsius)

# Graficar el scatter plot y la lnea de regresin
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
#plt.figure(figsize=(6, 4))

# Cargar la imagen de fondo
imagen_fondo = plt.imread('C:\\Users\\Maanas\\pictures\\fondo.jpg')
plt.imshow(imagen_fondo, extent=[np.min(celsius), np.max(celsius), np.min(fahrenheit), np.max(fahrenheit)], aspect='equal', alpha=0.3)

plt.scatter(fahrenheit,celsius, color='blue', label='Datos')

# El ploteo del predictor se ve super raro
plt.plot(prediccion,celsius, color='red', label='Regresin lineal')

print(prediccion)


```

```python
nueva_temp=np.array([-5.0], dtype=float)
prediccion = modelo.predict([nueva_temp])

plt.scatter(prediccion,nueva_temp, color='green', label='Regresin lineal')

print(prediccion)
```

```python
# Agregar etiquetas y ttulo
plt.title('Correlacin entre celsius y farenheit')
plt.xlabel('Celsius  (C)')
plt.ylabel('Farenheit')


# Mostrar la leyenda
plt.legend()

# Mostrar el scatter plot con la lnea de regresin
plt.show()
```
